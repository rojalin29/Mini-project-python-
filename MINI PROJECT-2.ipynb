```python
import sys
print("Welcome to Employee System")
l_id= input("Please Enter Login id:  ")
pwd = input("Please Enter Password:  ")
file = open("login.txt", "r")
re = file.readlines()
flag=0
for i in re:
    j = i.split()
    if l_id == j[0] and pwd == j[1]:
        if l_id == 'admin':
            flag = 1
            break
else:
    print("Invalid id and password")
    sys.exit()
def add_emp(emp_id, emp_name, emp_DOJ, emp_designation, emp_salary):
    file1 = open('Employee.txt', 'a')
    file2 = open('login.txt', 'a')
    emp = emp_id + ',' + emp_name + ',' + emp_DOJ + ',' + emp_designation + ',' + emp_salary + '\n'
    lst = emp_name.split()
    emp_login = emp_id + ' ' + lst[0]+'\n'
    file1.writelines(emp)
    file2.writelines(emp_login)
    file1.close()
    file2.close()
    return "Employee details are added"
def remove(emp_id, files, s):
    f = open(files, 'r')
    a = f.readlines()
    for i in range(len(a)):
        j = a[i].split(s)
        if j[0] == emp_id:
            a[i] = ''
            break
    f.close()
    f = open(files, 'w')
    f.writelines(a)
    f.close()
def remove_emp(emp_id):
    remove(emp_id, 'Employee.txt', ',')
    remove(emp_id, 'HR.txt', ',')
    remove(emp_id, 'login.txt', ' ')
def check(emp_id):
    temp = False
    file = open("Employee.txt", 'r')
    employee = file.readlines()
    for i in employee:
        j = i.split(',')
        if j[0] == emp_id:
            temp = True
    return temp
def add_hr(emp_id, hr_dept, hr_role):
    file = open('HR.txt', 'a')
    hr = emp_id + ',' + hr_dept + ',' + hr_role + '\n'
    file.writelines(hr)
    file.close()
    return "HR details are added"
def remove_hr(emp_id):
    remove(emp_id, 'HR.txt', ',')
if flag==1:
    while 1:
        print("Welcome admin!!")
        print("Enter 1 to add employee\nEnter 2 to remove employee\nEnter 3 to add hr\nEnter 4 to remove hr\nEnter q to ""exit")
        opt = input("Enter your choice: ")
        if opt == '1':
            emp_id = input("Enter employee id: ")
            emp_name = input("Enter employee name: ")
            emp_DOJ = input("Enter DOJ: ")
            emp_designation = input("Enter the designation: ")
            emp_salary = input("Enter the salary: ")
            print(add_emp(emp_id, emp_name, emp_DOJ, emp_designation, emp_salary))
        elif opt == '2':
            emp_id = input("Enter employee id:")
            remove_emp(emp_id)
        elif  opt == '3':
            emp_id = input("Enter employee id:")
            hr_dept = input("Enter hr dept: ")
            hr_role = input("Enter hr role: ")
            print(add_hr(emp_id, hr_dept, hr_role))
        elif opt == '4':
            emp_id = input("Enter employee id:")
            print("Is he/she leaving the company permanently?")
            y = input("Enter yes or no: ")
            if y == 'yes':
                remove_emp(emp_id)
            else:
                remove_hr(emp_id)
        elif opt == 'q':
            break
```
